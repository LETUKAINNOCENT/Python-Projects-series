```python
import yfinance as yf
import pandas as pd
```

```python
gbp_usd = yf.download("GBPUSD=X", period="2y", interval="1h")
eur_usd = yf.download("EURUSD=X", period="2y", interval="1h")
```

```python
df = pd.concat([gbp_usd[['Close']], eur_usd[['Close']]], axis=1)

df.columns = ['GBPUSD', 'EURUSD']
df = df.reset_index()

df.head()
```

```python
df.to_csv("gbpusd_h1_2year_data.csv")
```
