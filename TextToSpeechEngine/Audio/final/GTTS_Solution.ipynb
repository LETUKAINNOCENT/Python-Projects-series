```python
from gtts import gTTS
from pydub import AudioSegment
import re
```

```python
# 1. Read file
with open("output2.txt", "r", encoding="utf-8") as file:
    text = file.read()
```

```python
def clean_text(text):
    # Remove extra whitespace and empty lines
    text = re.sub(r'\s+', ' ', text)

    # Remove unwanted characters (keep basic punctuation)
    text = re.sub(r'[^a-zA-Z0-9.,!?\'" ]+', '', text)

    return text.strip()

cleaned_text = clean_text(text)
```

```python
# 3. Handle long text (gTTS limit workaround)
def split_text(text, max_length=4000):
    chunks = []
    while len(text) > max_length:
        split_index = text.rfind(' ', 0, max_length)
        chunks.append(text[:split_index])
        text = text[split_index:]
    chunks.append(text)
    return chunks

chunks = split_text(cleaned_text)
```

```python
for i, chunk in enumerate(chunks):
  print(f"{i} : {chunk}")
```

```python
chunks2 = chunks[24:]
for i, chunk in enumerate(chunks2):
  print(f"{i+24} : {chunk}")
```

```python
# 4. Convert each chunk to audio
for i, chunk in enumerate(chunks2):
    tts = gTTS(text=chunk, lang='en')
    tts.save(f"output_part_{i+24}.mp3")

print("Audio files generated successfully!")
```

```python

combined = AudioSegment.empty()

for i in range(16):
    audio = AudioSegment.from_mp3(f"output_part_{i}.mp3")
    combined += audio

combined.export("final_output2.mp3", format="mp3")
```
