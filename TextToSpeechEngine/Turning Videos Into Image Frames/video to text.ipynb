```python
import os
import cv2
import numpy as np
import pytesseract
from glob import glob
from fpdf import FPDF
from PIL import Image
from matplotlib import pyplot as plt
from reportlab.platypus import SimpleDocTemplate, Paragraph
from reportlab.lib.styles import getSampleStyleSheet
```

## make sure the path exists and then print the contents of the path

```python
def create_dir(path):
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
video_Folder_path = r"C:\Users\MIL\Videos\OBS"
output_path       = r"C:\Users\MIL\Documents\Python OutPut Images"

matches = glob(video_Folder_path)
if matches:
    for folder in matches:
        for item in glob(os.path.join(folder, "*")):
            video_path = item
video_path
```

## Working on the video

```python
cap = cv2.VideoCapture(video_path)

prev_gray = None
prev_img = None
last_score = 0
page_count = 0
frame_idx  = 0
threshold  = 1
```

```python
while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    if prev_gray is None:
        cv2.imwrite(os.path.join(output_path, f"page_{page_count}.png"), frame)
        print("Saved img ",page_count+1)
        page_count += 1
        prev_gray = gray
        prev_img = frame
        continue

    diff      = cv2.absdiff(prev_gray, gray)
    score     = diff.mean()
    prev_gray = gray
    
    if score > threshold:
        cv2.imwrite(os.path.join(output_path, f"page_{page_count}.png"), frame)
        print("Saved img ",page_count+1)

        page_count += 1
        frame_idx += 1
        prev_gray = gray
        prev_img = frame
        
cap.release()
```

## TURNING THE IMAGES INTO TEXT

```python
myconfig = r'--oem 3 --psm 6'
text  = ""
matches = glob(output_path)
    
cout = 0
all_text = ""
for image_file in os.listdir(output_path):
    
    image_path = os.path.join(output_path, f"page_{cout}.png")
    
    print(image_path)
    
    if os.path.exists(image_path):
        img = Image.open(image_path)
        text = pytesseract.image_to_string(img, config=myconfig)
    
        all_text += text 
    cout += 1 
```

```python
output_file_path = r"C:\Users\MIL\Documents\Python OutPut Files\textFiles"
create_dir(output_file_path)

with open(f"{output_file_path}/output.txt", "w", encoding="utf-8") as f:
    f.write(all_text)

print("Text file saved as output.txt")
```
